```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
face_cascade = cv2.CascadeClassifier('C:\\Users\\Mahesh\\Documents\\ArtificialIntelligence\\UntitledFolder1\\haarcascade_frontalface_default.xml')
```

```python
img=cv2.imread('test1.jpg')
```

```python
plt.imshow(img)
print(img.shape)
img=cv2.resize(img,(600,600))
print(img.shape)
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img)
print(gray_img.shape)
```

```python
test2=face_cascade.load('haarcascade_frontalface_default.xml')
```

```python
faces=face_cascade.detectMultiScale(gray_img, scaleFactor=1.01,minNeighbors=3)
```

```python
print(test2)
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x,y), (x+w,y+h), (255,255,255) ,3)
```

```python
plt.imshow(img)
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(gray_img, (x,y), (x+w,y+h), (255,0,0) ,3)
plt.imshow(gray_img)
```
